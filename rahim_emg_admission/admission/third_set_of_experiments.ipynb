```python
import ML_lib as ml
import pandas as pd
```

```python
n_estimator_gbc = 400
min_samples_split = 0.001
max_features = 0.5
verbose = True

DATA_LOC = '/Users/fatemehrahimian/Documents/DeepMedicine/Data/PR001/Exp_data/set3/'

frac = 0.5
period = ['6', '6_12', '12_24', '24_36', '36_48', '48_60']
k = 1
```

---

# Third set of experiments

This set of experiments evaluate Gradient Boosting for different study periods with both left and right censoring

```python
qa_lr_result_list = []
labels = []
for p in period:
    print ''
    print p
    filename = 'qa_' + p + '.csv'
    data_p = pd.read_csv(DATA_LOC + filename) 
    data_p = data_p.drop('Unnamed: 0', axis = 1)
    data_p = data_p.sample(frac = frac)
    data_p = ml.get_dummies(data_p)
    
    results = ml.GB_ER(data_p, k, n_estimator_gbc, min_samples_split, max_features, verbose)    
    qa_lr_result_list.append(results[2])
    labels.append(p)
    print '-----------'

print 'done!'

y_series_qa_lr = ml.plot_ROC_Over_Period(qa_lr_result_list, labels)
print y_series_qa_lr
```

```python
fr_lr_result_list = []
labels = []
for p in period:
    print ''
    print p
    filename = 'fr_' + p + '.csv'
    data_p = pd.read_csv(DATA_LOC + filename) 
    #data_p = data_p.drop('Unnamed: 0', axis = 1)
    data_p = data_p.sample(frac = frac)
    data_p = ml.get_dummies(data_p)
    
    results = ml.GB_ER(data_p, k, n_estimator_gbc, min_samples_split, max_features, verbose)    
    fr_lr_result_list.append(results[2])
    labels.append(p)
    print '-----------'

print 'done!'

y_series_fr_lr = ml.plot_ROC_Over_Period(fr_lr_result_list, labels)
print y_series_fr_lr
```
