```python
import ML_lib as ml
import pandas as pd
```

```python
k = 3
n_estimator_gbc = 800
n_estimator_rf = 50
min_samples_split = 0.001
max_features = 0.5
max_depth = 10
verbose = True

frac = 0.5

DATA_LOC = '/Users/fatemehrahimian/Documents/DeepMedicine/Data/PR001/Exp_data/set3/'
```

---

# First set of experiments

In the first set of experiments, we apply Logistic regression, Random Forest, and Gradient Boosting on QA data.

**Note:**  When preprocessing QA data, drop 'previous_admissions', but keep 'last_year_admissions', just as in the paper.


### QA

```python
filename = 'qa_24.csv'
qa = pd.read_csv(DATA_LOC + filename)
qa = qa.sample(frac = frac)
qa = qa.drop('previous_admissions', axis = 1)
qa = qa.drop('Unnamed: 0', axis = 1)
qa = ml.get_dummies(qa)

qa_gb_results = ml.GB_ER(qa, k, n_estimator_gbc, min_samples_split, max_features, verbose)
qa_rf_results = ml.RF_ER(qa, k, n_estimator_rf, min_samples_split, max_features, verbose)
qa_lr_results = ml.LR_ER(qa, k, verbose)
print 'done!'
```

```python
#ml.plot_ALL(k, qa_gb_results[2], 'GBC')

#ml.plot_ALL(k, qa_rf_results[2], 'RF')

#ml.plot_ALL(k, qa_lr_results[2], 'LR')

ml.plot_ROC_Multiple_CLF(k, [qa_gb_results[2], qa_rf_results[2], qa_lr_results[2]], ['GBC', 'RF', 'LR'])

ml.plot_Calibration_Plot_multiple_CLF(k, [qa_gb_results[2], qa_rf_results[2], qa_lr_results[2]], ['GBC', 'RF', 'LR'])

ml.plot_Calibration_Centiles_multiple_CLF(k, [qa_gb_results[2], qa_rf_results[2], qa_lr_results[2]], ['GBC', 'RF', 'LR'])

ml.plot_Precision_Recall_multiple_CLF(k, [qa_gb_results[2][0], qa_rf_results[2][0], qa_lr_results[2][0]], ['GBC', 'RF', 'LR'])
```
